В датасете представленны разничные характеристики телефонов (признаки) и их ценовая группа (столбец **price_range** -  целевая переменная). Необходимо спрогнозировать, к какому классу по стоимости относится телефон.

```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression as LogR
from sklearn.metrics import accuracy_score, f1_score, classification_report, roc_auc_score, confusion_matrix
from sklearn.preprocessing import MinMaxScaler, RobustScaler
```

```python
from google.colab import drive
drive.mount('/content/drive')
data = pd.read_csv('drive/My Drive/PhoneClassification.csv')
data.head()
```

```python
x = data[['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g', 'int_memory', 'm_dep', 'mobile_wt',
          'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g', 'touch_screen', 'wifi']]
y = data['price_range']

x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.3, random_state = 34)
log_res = LogR(solver='sag', max_iter=10000)
model = log_res.fit(x_train, y_train)
pred_train = model.predict(x_train)
pred_train
```

```python
pred_train_proba = model.predict_proba(x_train)
pred_train_proba
```

```python
pred_test = model.predict(x_test)
pred_test
```

```python
pred_test_proba = model.predict_proba(x_test)
pred_test_proba
```

## Метрики к модели мультиклассификации

```python
print('Metrics for train:\n ', classification_report(y_train, pred_train))
```

```python
print('Metrics for test:\n ', classification_report(y_test, pred_test))
```

## Собственные функции для подсчета метрик

```python
tp = sum((y_test == 1) & (pred_test == 1))
fp = sum((y_test == 0) & (pred_test == 1))
tn = sum((y_test == 0) & (pred_test == 0))
fn = sum((y_test == 1) & (pred_test == 0)) 
print(tp, fp, tn, fn)
```

```python
def new_accuracy(tp, fp, tn, fn):
  return (tp+tn)/(tp+tn+fp+fn)

new_accuracy(tp, fp, tn, fn)
```

```python
def new_precision(tp, fp):
  return tp/(tp+fp)

new_precision(tp, fp)
```

```python
def new_recall(tp, fn):
  return tp/(tp+fn)

new_recall(tp, fn)
```

```python
def new_f1(tp, fp, fn):
  return (2 * new_precision(tp, fp) * new_recall(tp, fn))/(new_precision(tp, fp) + new_recall(tp, fn))

new_f1(tp, fp, fn)
```

Функции выдали верный результат

## Модель с учетом нормализации и стандартизации

```python
min_max = MinMaxScaler()
min_max.fit(x_train)
x_train_min_max = min_max.transform(x_train)
x_test_min_max = min_max.transform(x_test)
```

```python
lr = LogR(solver = "sag", max_iter = 10000)
model_mm = lr.fit(x_train_min_max, y_train)
pred_min_max = model_mm.predict(x_test_min_max)

print(classification_report(y_test, pred_min_max))
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.6, random_state = 34)

min_max.fit(x_train)
x_train_min_max = min_max.transform(x_train)
x_test_min_max = min_max.transform(x_test)

model_mm = lr.fit(x_train_min_max, y_train)
pred_min_max = model_mm.predict(x_test_min_max)

print(classification_report(y_test, pred_min_max))
```

```python
robust = RobustScaler()
robust.fit(x_train)
x_train_robust = robust.transform(x_train)
x_test_robust = robust.transform(x_test)

model_rob = lr.fit(x_train_robust, y_train)
pred_rob = model_rob.predict(x_test_robust)

print(classification_report(y_test, pred_rob))
```
